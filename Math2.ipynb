```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Круг:

```python
r=3
x0=0
y0=0
x=np.linspace((x0-r),(r+x0),1000)
y1=[]
y2=[]

for i in x:
    y1.append((math.sqrt(r**2-(i-x0)**2))+y0)
    y2.append(-(math.sqrt(r**2-(i-x0)**2))+y0)
plt.plot(x,y1)
plt.plot(x,y2)
```

Овал:

```python
r1=3
r2=5
x0=0
y0=0
x=np.linspace((x0-r1),(r1+x0),1000)
y1=[]
y2=[]

for i in x:
    y1.append((math.sqrt(r2**2-((i-x0)**2)*(r2**2)/r1**2))+y0)
    y2.append(-(math.sqrt(r2**2-((i-x0)**2)*(r2**2)/r1**2))+y0)
plt.plot(x,y1)
plt.plot(x,y2)
```

Параллельные плоскости:

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax=Axes3D(fig)
X=np.arange(-5,5,0.5)
Y=np.arange(-5,5,0.5)
X, Y = np.meshgrid(X,Y)
Z = 2*X + 5*Y
A=np.arange(-5,5,0.5)
B=np.arange(-5,5,0.5)
A, B = np.meshgrid(A,B)
C = 2*A + 5*B + 50

ax.plot_wireframe(X,Y,Z)
ax.plot_wireframe(A,B,C)
ax.scatter(0,0,0,'z',50,'red')
show()
```

Поверхности второго порядка:

```python
fig=figure()
ax=Axes3D(fig)
X=np.arange(-5,5,0.5)
Y=np.arange(-5,5,0.5)
X, Y = np.meshgrid(X,Y)
Z = np.sqrt(X**2+Y**2+5)
Z2 = -np.sqrt(X**2+Y**2+5)

ax.plot_wireframe(X,Y,Z)
ax.plot_wireframe(X,Y,Z2)

ax.scatter(0,0,0,'z',50,'red')
show()
```
